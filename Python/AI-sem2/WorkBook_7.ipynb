Задание 

Реализовать классы нейросетей по аналогии с классом OurNeuralNetwork.  
Данные нейросети:  
- три входа (x<sub>1</sub>, x<sub>2</sub>, x<sub>3</sub>)
- три нейрона в скрытых слоях (h<sub>1</sub>, h<sub>2</sub>, h<sub>3</sub>)
- выход (o<sub>1</sub>)

Нейроны имеют идентичные веса и пороги:
- w = [0.5, 0.5, 0.5]
- b = 0

Данные нейросети:
- два входа (x<sub>1</sub>, x<sub>2</sub>)
- два нейрона в скрытых слоях (h<sub>1</sub>, h<sub>2</sub>)
- два выхода (o<sub>1</sub>, o<sub>2</sub>)

Нейроны имеют идентичные веса и пороги:
- w = [1, 0]
- b = 1

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

class Neuron:
    def __init__(self, weights, bias):
        self.weights = weights
        self.bias = bias
    def feedforward(self, inputs):
        total = np.dot(self.weights, inputs) + self.bias
        return sigmoid(total)


# 1 пример нейросети
class NeuralNetwork_1:
    def __init__(self):
        weights = np.array([0.5, 0.5, 0.5])
        bias = 0
        self.h1 = Neuron(weights, bias)
        self.h2 = Neuron(weights, bias)
        self.h3 = Neuron(weights, bias)
        self.o1 = Neuron(weights, bias)
    def feedforward(self, x):
        out_h1 = self.h1.feedforward(x)
        out_h2 = self.h2.feedforward(x)
        out_h3 = self.h3.feedforward(x)
        out_o1 = self.o1.feedforward(np.array([out_h1, out_h2, out_h3]))
        return out_o1
    
# 2 пример нейросети
class NeuralNetwork_2:
    def __init__(self):
        weights = np.array([1, 0])
        bias = 1
        self.h1 = Neuron(weights, bias)
        self.h2 = Neuron(weights, bias)
        self.o1 = Neuron(weights, bias)
    def feedforward(self, x):
        out_h1 = self.h1.feedforward(x)
        out_h2 = self.h2.feedforward(x)
        out_o1 = self.o1.feedforward(np.array([out_h1, out_h2]))
        out_o2 = self.o1.feedforward(np.array([out_h1, out_h2]))
        return out_o1, out_o2
    
example_1 = NeuralNetwork_1()
example_2 = NeuralNetwork_2()

x_1 = np.array([2, 3, -1])
x_2 = np.array([2, 3])

print(example_1.feedforward(x_1))
print(example_2.feedforward(x_2))
```

Реализуйте классы нейронных сетей с использованием других функций активации.



```python
import numpy as np


# 1 пример нейросети с сигмоидой
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

class Neuron_sigmoid:
    def __init__(self, weights, bias):
        self.weights = weights
        self.bias = bias
    def feedforward(self, inputs):
        total = np.dot(self.weights, inputs) + self.bias
        return sigmoid(total)

class NeuralNetwork_1:
    def __init__(self):
        weights = np.array([0.5, 0.5, 0.5])
        bias = 0
        self.h1 = Neuron_sigmoid(weights, bias)
        self.h2 = Neuron_sigmoid(weights, bias)
        self.h3 = Neuron_sigmoid(weights, bias)
        self.o1 = Neuron_sigmoid(weights, bias)
    def feedforward(self, x):
        out_h1 = self.h1.feedforward(x)
        out_h2 = self.h2.feedforward(x)
        out_h3 = self.h3.feedforward(x)
        out_o1 = self.o1.feedforward(np.array([out_h1, out_h2, out_h3]))
        return out_o1
    

# 2 пример нейросети с гиперболическим тангенсом
def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

class Neuron_tanh:
    def __init__(self, weights, bias):
        self.weights = weights
        self.bias = bias
    def feedforward(self, inputs):
        total = np.dot(self.weights, inputs) + self.bias
        return tanh(total)


class NeuralNetwork_2:
    def __init__(self):
        weights = np.array([1, 0])
        bias = 1
        self.h1 = Neuron_tanh(weights, bias)
        self.h2 = Neuron_tanh(weights, bias)
        self.o1 = Neuron_tanh(weights, bias)
    def feedforward(self, x):
        out_h1 = self.h1.feedforward(x)
        out_h2 = self.h2.feedforward(x)
        out_o1 = self.o1.feedforward(np.array([out_h1, out_h2]))
        out_o2 = self.o1.feedforward(np.array([out_h1, out_h2]))
        return out_o1, out_o2
    

# 3 пример нейросети с Relu
def relu(x):
    return np.maximum(0, x)

class Neuron_Relu:
    def __init__(self, weights, bias):
        self.weights = weights
        self.bias = bias
    def feedforward(self, inputs):
        total = np.dot(self.weights, inputs) + self.bias
        return relu(total)


class NeuralNetwork_3:
    def __init__(self):
        weights = np.array([1, 0])
        bias = 1
        self.h1 = Neuron_Relu(weights, bias)
        self.h2 = Neuron_Relu(weights, bias)
        self.o1 = Neuron_Relu(weights, bias)
    def feedforward(self, x):
        out_h1 = self.h1.feedforward(x)
        out_h2 = self.h2.feedforward(x)
        out_o1 = self.o1.feedforward(np.array([out_h1, out_h2]))
        out_o2 = self.o1.feedforward(np.array([out_h1, out_h2]))
        return out_o1, out_o2
    

# Примеры использования простых нейросетей
example_1 = NeuralNetwork_1()
example_2 = NeuralNetwork_2()
example_3 = NeuralNetwork_3()

x_1 = np.array([2, 3, -1])
x_2_3 = np.array([2, 3])

print(example_1.feedforward(x_1))
print()
print(example_2.feedforward(x_2_3))
print()
print(example_3.feedforward(x_2_3))
```
